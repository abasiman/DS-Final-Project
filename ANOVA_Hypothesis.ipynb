```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import category_encoders as ce
from kmodes.kmodes import KModes
```

```python
df=pd.read_csv("dataset/subscriptions_60000_utf8.csv")
```

```python
#unique values in 'revenue' for each service
for service in df['service'].unique():
    unique_values = df.loc[df['service'] == service, 'revenue'].unique()
    print(f"Service {service}: {unique_values}")
```

```python
import pandas as pd
from scipy.stats import f_oneway


top_services = df.groupby('service')['revenue'].nunique().sort_values(ascending=False).head(5).index


df_top = df[df['service'].isin(top_services)]

# Perform ANOVA
f_statistic, p_value = f_oneway(*[df_top['revenue'][df_top['service'] == service] for service in top_services])


print(f'F-statistic: {f_statistic}')
print(f'P-value: {p_value}')

# Resul Intrepretations
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in revenue across different services.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in revenue across different services.")
```
